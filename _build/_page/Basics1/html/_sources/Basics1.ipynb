## ユークリッドの互除法
- ここで、プログラミングらしい課題としてユークリッドの互除法というアルゴリズムの実装にトライしてみましょう。
-  ユークリッドの互除法とは、2つの整数、aとbの最大公約数を求めるアルゴリズムです。例えば、2485と1162の最大公約数を求めたい場合

$2485 \div1162 = 2 余り 161$ 

$1162 \div 161 = 7 余り 35$ 

$161 \div 35 = 4 余り 21$ 

$35 \div 21 = 1 余り 14$

$21 \div 14 = 1 余り 7$

$14 \div 7 = 2 余り 0$

割る数と余りを次回の割られる数と割る数にして、次々と計算し、最後に割り切れたら終わりで、最後の割る数が最大公約数です。



## 問題3
### 例題
- ユークリッドの互除法をPythonで実装しましょう。ヒントwhileを用います。余りがゼロでないならば、〜を実行せよ、という形式をとります。また、余りを計算するには%演算子を使います。

````python
#答え
a = 2485
b = 1162
while b !=0:
    r = a % b#入れ替え用のrを用意する 
    a = b
    b = r
    print(a,b)
````

````python
#別解
a = 2485
b = 1162
while b !=0:
    a, b = b, a % b#この書き方で複数の変数に代入できて、入れ替え用を用意する必要もない。 
    print(a,b)
````

### 問題 3-1
与えられた整数nに対して、nの階乗（n!）を求める関数factorialを定義してください。ただし、0の階乗は1と定義します。

````python
def factorial(n):
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n+1):
            result *= i
        return result

````

````python
# 例: 5の階乗を計算
result = factorial(5)
result
````

与えられた整数nが素数かどうかを判定する関数is_primeを作成してください。



````python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

n = 17
result = is_prime(n)
````

### 問題3-3
フィボナッチ数列の要素を生成し続けます。要素が1000を超えるような最初のフィボナッチ数を求めてください。
$$F_0 = 0$$
$$F_1 = 1$$
$$F_{n+2} = F_n + F_{n+1} (n ≥ 0)$$

````python
def first_fib_over_n(n):
    a, b = 0, 1
    while True:
        a, b = b, a + b
        if b > n:
            break
    return b

n = 1000
result = first_fib_over_n(n)
````

````python
result
````

## いろいろなプログラミングテクニック

#### forloopの進んだ使い方。enumerateとzip
- リストをループしているときにそのインデックスもあわせて取得したい場合があります。
- enumerate()を使ってループすると、インデックスと要素を使ってループすることができます。

````python
l = ['エミール', 'ゲオルグ','マックス', 'タルコット', 'ロバート']
for idx, name in enumerate(l):
    print("私（{}）のインデックスは{}番です".format(name,idx))
````

- zip()を使うとペアごとにループできます。

````python
l = ['エミール', 'ゲオルグ','マックス', 'タルコット', 'ロバート']
m = ['デュルケム','ジンメル','ウェーバー','パーソンズ','マートン']
for first,last in zip(l,m):
    print("私は{}・{}です".format(first,last))
````

#### リスト内包表記

- [item for item in list]のような形でリストの内部でfor文を使う方法です。pythonプログラミングで非常によく使われます。


````python
nums = [2, 4, 5, 6, 8, 10, 12]
new_nums = [num // 2 for num in nums]
print(new_nums)
````

```{important}
- さらにifで条件付けることができます。
  - ifのみの場合はforの後にifをおきます。


````python
nums = [2, 4, 5, 6, 8, 10, 12]
new_nums = [num // 2 for num in nums if num > 5]
print(new_nums)
````

```{important}
- if else で条件付けるときはforの前にif...else をおきます。

````python
nums = [2, 4, 5, 6, 8, 10, 12]
new_nums = [num // 2 if num > 5 else num for num in nums]
print(new_nums)
````

````python
d = {'マックス': 1864, 'エミール': 1858, 'ゲオルグ': 1858, 'タルコット': 1902, 'ロバート': 1910}#書き込み練習用の辞書を作成
````

- withを使います。

````python
with open('sample.csv','w') as f:
  for key, value in d.items():
    f.write(f"{key},{value}\n")
````

````python
%ls#sample.csvができている
````

````python
%cat sample.csv#中身を確認
````

- withを使わずにも書けますが、ファイルオブジェクトをクローズする必要があります。
  - hoge.close()とします。


````python
f = open('sample.csv','w')
for key, value in d.items():
  f.write(f"{key},{value}\n")
f.close()
````

````python
%cat sample.csv#中身を確認
````

- 'a'で開くと上書きモードとなる。

````python
with open('sample.csv','a') as f:
  key = "ニクラス"
  value =  1927
  f.write(f"{key},{value}\n")
````

````python
%cat sample.csv#中身を確認
````

### open()による読み込み
- open()を使って"r"読み込みモードで開く。


```{important}

- Windowsでは文字コードが異なるので、encoding = 'utf-8'とする。


```{important}
- 読み込みのメソッドは複数会ってそれぞれ挙動が異なるので注意してください。
- read()
- readlines()
- readline()

- read()メソッドでは、ファイルが一つの文字列として読みこまれる。


````python
with open('sample.csv','r',encoding = 'utf-8') as f:
  sample = f.read()
sample
````

- readlines()は改行コード(\n)で区切ったリストを返す。


````python
with open('sample.csv','r',encoding = 'utf-8') as f:
  sample = f.readlines()
sample
````

- readline()はファイルを一行ずつ呼んでいきます。
  - whileの無限ループを使って次々に呼び出していき、最後まで読むと、空の文字列を返すので、if notで条件づけてbreakで無限ループを脱出する。


```{note}注意
- ファイルオブジェクトは１回呼び出すと中身が更新される（例えば、f.read()をすると、fの中身は空になります。）
```

````python
with open('sample.csv','r',encoding = 'utf-8') as f:
  while True:#一行ずつ読み込んでいき、空になったらbreak
      line = f.readline()
      print(line)
      if not line:
          break
````

#### csvによる書き込み
- csv.writer()で書き込み用オブジェクトを作成します。

```{important}
- 書き込みのメソッドもwriterow(), writerows()の区別があります。

- writerow()メソッドを用いて、一行ずつ書き込む。


````python
sample = [['マックス','1864'], ['エミール', '1858'], ['ゲオルグ', '1858']]
import csv
with open('sample.csv','w') as f:
    writer = csv.writer(f)
    for row in sample:
        writer.writerow(row)
````

````python
%cat sample.csv
````

- wroterows()は二次元リストを一気に書き込む。


````python
with open('sample.csv','w') as f:
    writer = csv.writer(f)
    writer.writerows(sample)
 
````

````python
%cat sample.csv
````

### csvによる読み込み
- reader()メソッドでファイルを読み込む。
- for文で一行ずつ取り出す。


````python
sample = []
with open('sample.csv','r') as f:
    reader = csv.reader(f)
    #print(type(reader))
    for item in reader:
        sample.append(item)
sample
````

```{note}
- csv.reader()で作られたreaderオブジェクトはイテレータと呼ばれるオブジェクトで、for 文で一個ずつ繰り返し呼び出して中身を取り出すことができる。

### pickleで読み書きする
- pickleはPythonオブジェクトのままバイナリ形式で保存し、読み込む。処理がPythonのみで完結している際は推奨。

```{note}
- pickleを使う場合には、with open()を用います。
- 'wb'はバイナリオブジェクトの書き込み
- 'rb'はバイナリオブジェクトの読み込み

- 書き込みはpickle.dump()を使う。


````python
sample = [['マックス','1864'], ['エミール', '1858'], ['ゲオルグ', '1858']]

import pickle
with open ('sample.pickle','wb') as f:#バイナリオブジェクトなので'w'ではなく'wb'とする。
  pickle.dump(sample,f)
````

````python
%ls
````

````python
%cat sample.pickle
````

- 読み込みはpickle.load()を使う。

````python
with open ('sample.pickle','rb') as f:#バイナリオブジェクトなので'r'ではなく'rb'とする。
  sample = pickle.load(f)
````
