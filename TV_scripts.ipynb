```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
url =f'https://subslikescript.com/series/Suits-1632701'#suits スクリプトページ
response = requests.get(url)
soup=BeautifulSoup(response.content,"html.parser")
```

```python
#中身をみて、エピソード名"Pilot"で検索してあたりをつける
soup
```

```python
if not os.path.exists('scripts'):
                os.makedirs('scripts')
#このあたりにある
#<div class="season"><h4>Season 1</h4><ul><li>1. <a href="/series/Suits-1632701/season-1/episode-1-Pilot" title='Transcript of episode 1 "Pilot"'>Pilot</a></li><li>2
seasons = [item for item in soup.find_all('div', class_="season")]
for season in seasons:
    hrefs = [item['href'] for item in season.find_all("a", href=True)]#hrefが存在するaタグ（リンクurl）を抜き出す。
    for href in hrefs:
        response = requests.get("https://subslikescript.com"+href)#完全なurlにしてrequests
        soup_href = BeautifulSoup(response.content,"html.parser")
        scripts = soup_href.find("div", class_="full-script")#full-scriptに本文がある。
        for script in scripts(["script", "ins"]):# 広告が入っているため、スクリプトとインスタンスタグを削除
            script.decompose()#元のsoup_href()から指定されたオブジェクトを削除する
        clean_text = scripts.get_text(separator="\n")
        with open(f'scripts/'+href.split('/')[4], 'w') as f:#hrefの5番目の/の後にエピソード名があるのでファイル名とする
                f.write(clean_text)
                print(f"Downloaded: " +href.split('/')[4] +'.txt')






```

```python
# ディレクトリへのパス
directory = 'scripts/'

# ディレクトリ内のファイル名リストを取得
names = [item for item in os.listdir(directory)]

# ファイル名を変更
for name in names:
    old_path = os.path.join(directory, name)
    new_path = old_path + '.txt'
    os.rename(old_path, new_path)
```
